```python
# Import pandas for data analysis
import pandas as pd
```

```python
# Load the dataset into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/hospital data analysis.csv')
```

```python
# Display dataset structure and data types
# Purpose: Understand columns, data types, and check for missing values
df.info()
```

```python
# Generate statistical summary of numerical columns
# Purpose: Understand distribution, mean, min, max, etc.
df.describe()
```

```python
# Analyze frequency of each medical condition
# Purpose: Identify the most common illnesses among patients
condition_counts = df['Condition'].value_counts()
print(condition_counts)

# Visualization: Bar chart of most common conditions
# Purpose: Provide a visual representation of condition frequency
import matplotlib.pyplot as plt

condition_counts.plot(kind='bar')
plt.title("Most Common Medical Conditions")
plt.xlabel("Condition")
plt.ylabel("Number of Patients")
plt.show()
```

```python
# Calculate the average age of patients
# Purpose: Understand the general age group of patients
average_age = df['Age'].mean().round(0)

print(average_age)
```

```python
# Count number of patients by gender
# Purpose: Analyze gender distribution in the dataset
gender_distribution = df['Gender'].value_counts()

print(gender_distribution)

df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.show()
```

```python
# Calculate the average treatment cost
# Purpose: Understand general cost level of treatments
average_cost = df['Cost'].mean().round(2)

print(average_cost)
```

```python
# Calculate average cost per condition
# Purpose: Identify which conditions are most expensive to treat
cost_by_condition = df.groupby('Condition')['Cost'].mean().round(2).sort_values(ascending=False)

print(cost_by_condition)

cost_by_condition.plot(kind='bar')
plt.title("Average Cost by Condition")
plt.show()
```

```python
# Calculate average hospital stay per condition
# Purpose: Identify conditions requiring longer hospitalization
length_of_stay = df.groupby('Condition')['Length_of_Stay'].mean().round(0).sort_values(ascending=False)

print(length_of_stay)
```

```python
# Count how many patients were readmitted
# Purpose: Evaluate hospital performance and treatment effectiveness
readmission_counts = df['Readmission'].value_counts()

print(readmission_counts)
```

```python
# Analyze readmission frequency per condition
# Purpose: Identify conditions with higher chances of readmission
readmission_by_condition = df.groupby('Condition')['Readmission'].value_counts()

print(readmission_by_condition)
```

```python
# Calculate average patient satisfaction score
# Purpose: Measure overall patient experience
average_satisfaction = df['Satisfaction'].mean().round(2)

print(average_satisfaction)
```

```python
# Analyze satisfaction level by condition
# Purpose: Identify which conditions have better or worse patient experiences
satisfaction_by_condition = df.groupby('Condition')['Satisfaction'].mean().round(2).sort_values(ascending=False)

print(satisfaction_by_condition)

satisfaction_by_condition.plot(kind='bar')
plt.title("Patient Satisfaction by Condition")
plt.show()
```

```python
# Correlation Analysis
# Purpose: Identify relationships between numerical variables
correlation_matrix = df[['Age', 'Cost', 'Length_of_Stay', 'Satisfaction']].corr()

print(correlation_matrix)


plt.imshow(correlation_matrix)
plt.colorbar()
plt.title("Correlation Matrix")
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.show()
```

```python
# Group Comparisons

# Cost by Gender
cost_by_gender = df.groupby('Gender')['Cost'].mean().round(2)
print("\nAverage Cost by Gender:\n", cost_by_gender)

# Satisfaction vs Readmission
satisfaction_readmission = df.groupby('Readmission')['Satisfaction'].mean().round(2)
print("\nSatisfaction by Readmission:\n", satisfaction_readmission)

# Average Age by Condition
age_by_condition = df.groupby('Condition')['Age'].mean().round(0).sort_values(ascending=False)
print("\nAverage Age by Condition:\n", age_by_condition)
```
